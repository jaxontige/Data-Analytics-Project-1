```python
#Imports

import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Specify the path to your CSV file
csv_file_path5 = 'Table26-2019-by-Month.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path5)

# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Print the DataFrame
df
```

# Average Mortgage Rate  

```python
# Extract relevant columns and filter data for years 2006-2010
df_home_prices = df.loc[(df['Year'] >= 2006) & (df['Year'] <= 2010), ['Year', 'Month', 'Contract Interest Rate (%)']]

# Convert 'Year' and 'Month' to integers
df_home_prices['Year'] = df_home_prices['Year'].astype(int)
df_home_prices['Month'] = df_home_prices['Month'].astype(int)

# Combine 'Year' and 'Month' columns to create a 'Month-Year' column
df_home_prices['Month-Year'] = df_home_prices['Year'].astype(str) + '-' + df_home_prices['Month'].astype(str)

# Convert 'Month-Year' to datetime format
df_home_prices['Month-Year'] = pd.to_datetime(df_home_prices['Month-Year'], format='%Y-%m')

# Calculate monthly Contract Interest Rate
avg_home_prices = df_home_prices.groupby('Month-Year')['Contract Interest Rate (%)'].mean()
```

```python
# Plotting
plt.figure(figsize=(15, 6))

# Plotting mortgage rates
plt.plot(avg_home_prices, label="Average Mortgage Rate", color='b', marker='o')

# Labels and title
plt.xlabel("Month-Year")
plt.ylabel("Average Rate (%)")
plt.title("Average Mortgage Rates by Months (2006-2010)")
plt.xticks(rotation=45, ha='right') 
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# Average Home Price

```python
# Extract relevant columns and filter data for years 2006-2010
df_home_prices = df.loc[(df['Year'] >= 2006) & (df['Year'] <= 2010), ['Year', 'Month', 'Purchase Price ($thou)']]

# Convert 'Year' and 'Month' to integers
df_home_prices['Year'] = df_home_prices['Year'].astype(int)
df_home_prices['Month'] = df_home_prices['Month'].astype(int)

# Combine 'Year' and 'Month' columns to create a 'Month-Year' column
df_home_prices['Month-Year'] = df_home_prices['Year'].astype(str) + '-' + df_home_prices['Month'].astype(str)

# Convert 'Month-Year' to datetime format
df_home_prices['Month-Year'] = pd.to_datetime(df_home_prices['Month-Year'], format='%Y-%m')

# Calculate monthly average home prices
avg_home_prices = df_home_prices.groupby('Month-Year')['Purchase Price ($thou)'].mean()
```

```python
# Plotting
plt.figure(figsize=(15, 6))

# Plotting home prices
plt.plot(avg_home_prices, label="Average Home Price", color='y', marker='o')

# Labels and title
plt.xlabel("Month-Year")
plt.ylabel("Average Price ($thou)")
plt.title("Average Home Prices by Months (2006-2010)")
plt.xticks(rotation=45, ha='right') 
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
