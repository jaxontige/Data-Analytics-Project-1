```python
# Load the CSV file 
federal_rates_2006 = pd.read_csv("Source Data/Treasury Rates/2006-daily-treasury-rates.csv")

# Display sample data
federal_rates_2006.head()
```

```python
#Convert the date from a string to a date
federal_rates_2006['Date'] = pd.to_datetime(federal_rates_2006['Date'])

#Create a new month column
federal_rates_2006["Month"] = federal_rates_2006["Date"].dt.month
federal_rates_2006.head()
```

```python
#2006

#Separate just the 30 year loan column, with date and month column
thirty_yr_rates_06 = federal_rates_2006[["Date", "30 Yr", "Month"]]
renamed_06 = thirty_yr_rates_06.rename(columns={
                                '30 Yr' : '30 Yr 2006'})
                                
#cleaned_06 = renamed_06.dropna()
#cleaned_06

#Groupby Month
monthly_group_06 = round(renamed_06.groupby(['Month'])["30 Yr 2006"].mean(), 2).reset_index()
monthly_group_06

```

```python
#Merge seprate dataframes into one dataframe
merged_df = pd.merge(monthly_group_06, monthly_group_07, on='Month')
merged2_df = pd.merge(merged_df, monthly_group_08, on='Month')  
merged3_df = pd.merge(merged2_df, monthly_group_09, on='Month') 
merged4_df = pd.merge(merged3_df, monthly_group_10, on='Month') 

print(merged4_df)
```

```python
#Export dataframe to csv
merged4_df.to_csv('Output_Data/2005-2010, 30 Year Rates.csv', index=False)
```

```python
#Convert Dataframe to a vertical running list with all interest rates in one column
monthly_group_06["Year"]=2006
monthly_group_07["Year"]=2007
monthly_group_08["Year"]=2008
monthly_group_09["Year"]=2009
monthly_group_10["Year"]=2010

monthly_group_06.columns = ["Month", "30 Year", "Year"]
monthly_group_07.columns = ["Month", "30 Year", "Year"]
monthly_group_08.columns = ["Month", "30 Year", "Year"]
monthly_group_09.columns = ["Month", "30 Year", "Year"]
monthly_group_10.columns = ["Month", "30 Year", "Year"]

vertical = pd.concat([monthly_group_06, monthly_group_07, monthly_group_08, monthly_group_09, monthly_group_10]).reset_index(drop=True)

vertical.head()


```

```python
#Export to csv
vertical.to_csv('Output_Data/2005-2010, 30 Year Rates, Comparable.csv', index=False)
```

```python
#Convert month column to having two digits
vertical["two_digit_month"] = vertical["Month"].astype(str).str.zfill(2)
vertical.head()
```

```python
#Convert Year to a string
vertical["Year"]=vertical["Year"].astype(str)

#Create a new column with the combined month and year
vertical['Date']= vertical["two_digit_month"] + ', ' + vertical["Year"]
vertical.head()
```

```python
#Select Date and Interest Rate Columns
organized = vertical[['Date', '30 Year']]
organized.head()
```

```python
#Export to csv
formated_30_year_treasury = organized.to_csv("Output_Data/2005-2010, 30 Year Treasury Rates, Final.csv", index=False)
```

```python
#Create figure of treasury data
organized.plot(x='Date', y='30 Year', kind='line', color='skyblue')
plt.xlabel('Date')
plt.ylabel('Treasury Rate')
plt.title('Average 30 Year Treasury Rate')
plt.legend({'Treasury'})

#Save figure
plt.savefig("Output_Data/30 Year Treasury.png")

plt.show()
```
